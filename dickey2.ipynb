```python
def dickey_by_year(df):
    
    results=[]
    
    for column in df.columns:
        
        test=adfuller(df[column])
        adf_stat=test[0]
        p_val=test[1]
        crit=test[4]
        
        results.append({
            'Column': column,
            'ADF Statistic': adf_stat,
            'p-value': p_val,
            'Critical Value (1%)': crit.get('1%'),
            'Critical Value (5%)': crit.get('5%'),
            'Critical Value (10%)': crit.get('10%')
        })
        
    result_df=pd.DataFrame(results)
    
    return result_df
```

```python
def dickey_by_month(df):
    
    results=[]
    exclude=[]
    exclude.extend(['Month'])
    for month, group in df.groupby('Month'):
        for column in group.columns:
            if column not in exclude:
                adf_result = adfuller(group[column].dropna())
                results.append({
                'Month': month,
                'Column': column,
                'ADF Statistic': adf_result[0],
                'p-value': adf_result[1],
                'Critical Value (1%)': adf_result[4].get('1%'),
                'Critical Value (5%)': adf_result[4].get('5%'),
                'Critical Value (10%)': adf_result[4].get('10%')
                })
    results_df = pd.DataFrame(results)
    
    return results_df
```
